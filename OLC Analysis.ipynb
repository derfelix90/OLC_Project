```python
import pandas as pd
```

```python
loc = r"\\Mac\Home\Downloads\flights_table_DE"
all_flights = pd.read_table(loc)
all_flights = all_flights.drop("Unnamed: 0", axis =1)
```

```python
all_flights
```

```python
airfield_filter = all_flights.loc[all_flights["Airfield"].str.contains("Nastaetten")]

airfield_filter.head()
```

```python

airfield_filter.loc[:,["Pilot1"]] = [row.split("(",1)[0]for row in airfield_filter["Pilot"]]
airfield_filter.loc[:,["Pilot2"]] = airfield_filter["Pilot"].apply(lambda x: x.split(")",1)[1])
airfield_filter.loc[:,["Pilot2"]] = airfield_filter["Pilot2"].apply(lambda x: x.split("(",1)[0])
airfield_filter
```

```python
#Speed Ranking
pilot_perf = airfield_filter.groupby(["Pilot1"])["Speed"].mean().reset_index().sort_values(["Speed"],ascending =False)
pilot_perf
```

```python
#Points Ranking
pilot_perf = airfield_filter.groupby(["Pilot1"])["Points"].mean().reset_index().sort_values(["Points"],ascending =False)
pilot_perf
```

```python
airfield_filter.loc[:,["Date"]] = pd.to_datetime(airfield_filter["Date"])
airfield_filter
```

```python
airfield_filter["Year"] = [row.year for row in airfield_filter["Date"]]
airfield_filter
```

```python
#Pilot Speed Performance by year
pperf_byyear = airfield_filter.groupby(["Pilot1","Year"])["Speed"].mean()
pperf_byyear
```
