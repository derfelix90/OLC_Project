```python
import pandas as pd
Flights_1205 = pd.read_html("https://www.onlinecontest.org/olc-3.0/gliding/daily.html?sc=&st=olc&rt=olc&df=2018-05-12&c=C0&d-1987472-p=&sp=2018&paging=100000", header=0)
Flights_1205
```

```python
Flights_1205_df = Flights_1205[0]
Flights_1205_df
```

```python
Flights_1205_df.loc[:,"Punkte"]=Flights_1205_df.iloc[:,1].str.split(" ",expand = True)[0]
Flights_1205_df.loc[:,"kmh"]= Flights_1205_df.iloc[:,1].str.split(" ",expand = True)[2]
Flights_1205_df.loc[:,"Punkte"] = Flights_1205_df.loc[:,"Punkte"].str.replace(",",".")
Flights_1205_df.loc[:,"kmh"] = Flights_1205_df.loc[:,"kmh"].str.replace(",",".")
Flights_1205_df.loc[:,"Punkte"] = Flights_1205_df.loc[:,"Punkte"].astype(float)
Flights_1205_df.loc[:,"kmh"] = Flights_1205_df.loc[:,"kmh"].astype(float)
filtered_GER = Flights_1205_df["Startplatz"].str.contains("\(DE")
Flights_1205_df_GER = Flights_1205_df[filtered_GER]
Flights_1205_df_GER
```

```python
#airfields aus dguv Tabelle
import numpy as np
loc2 = r"\\Mac\Home\Documents\Airfields.xlsx"
airfields = pd.read_excel(loc2,names = ["Kurz","Breite","Länge","Langname","Symbol"])
#airfields["Kurz"] = airfields["Kurz"].str.strip(",")
airfields["Langname"] = airfields["Langname"].str.strip(",")
airfields["Kurz"]=airfields["Langname"].str[:10]
airfields.loc[airfields["Kurz"].str.contains("\("),["Kurz"]] = airfields["Kurz"].str.split("\(",expand=True)[0]
airfields.loc[airfields["Kurz"].str.contains("\/"),["Kurz"]] = airfields["Kurz"].str.split("\/",expand=True)[0]
airfields.loc[airfields["Kurz"].str.contains("\-"),["Kurz"]] = airfields["Kurz"].str.split("\-",expand=True)[0]

airfields["Kurz"] = airfields["Kurz"].str.replace("ä","ae")
airfields["Kurz"] = airfields["Kurz"].str.replace("ü","ue")
airfields["Kurz"] = airfields["Kurz"].str.replace("ö","oe")
airfields["Kurz"] = airfields["Kurz"].astype(str)
airfields["Kurz"] = airfields["Kurz"].drop_duplicates()
#ersetze B- durch Bad
#airfields.iloc[40:61,0] = airfields["Kurz"].str.replace("B-", "Bad ")


#Breite modifizieren
airfields["Breite"] = airfields["Breite"].astype(str)
airfields["Breite"]= airfields["Breite"].str.strip(",")
airfields.loc[airfields["Breite"].str.isnumeric(),["Breite"]] = airfields["Breite"].str[:2] + "."+ airfields["Breite"].str[2:]

#Länge modifizieren
airfields["Länge"] = airfields["Länge"].astype(str)
airfields["Länge"]= airfields["Länge"].str.strip(",")
#hier weiter - muss unterschieden werden, wo der Punkt zu setzen ist. wenn erste Zahl =1 dann 2 Stellen, sonst 1 Stelle
airfields.loc[(airfields["Länge"].str.isnumeric())&(airfields["Länge"].str[0]!="1"),["Länge"]] = airfields["Länge"].str[:1] + "."+ airfields["Länge"].str[1:]
airfields.loc[(airfields["Länge"].str.isnumeric())&(airfields["Länge"].str[0]=="1"),["Länge"]] = airfields["Länge"].str[:2] + "."+ airfields["Länge"].str[2:]
airfields = airfields.dropna(axis = 0, how="any")
airfields.head(50)
```

```python
for f in airfields["Kurz"]:
    try:
        Flights_1205_df_GER.loc[Flights_1205_df_GER.Startplatz.str.contains(f),"Länge"] =  airfields.loc[airfields["Kurz"]==f,"Länge"].values
        #print(airfields.loc[airfields["Kurz"]==f,"Länge"].values.astype(float))
    except TypeError:
        pass
for f in airfields["Kurz"]:
    try:
        Flights_1205_df_GER.loc[Flights_1205_df_GER.Startplatz.str.contains(f),"Breite"] =  airfields.loc[airfields["Kurz"]==f,"Breite"].values
    except TypeError:
        pass
    
Flights_1205_df_GER
```
